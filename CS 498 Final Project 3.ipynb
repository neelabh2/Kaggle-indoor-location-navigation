```python
# copy from https://github.com/location-competition/indoor-location-competition-20/blob/master/io_f.py

@dataclass
class ReadData:
    wifi: np.ndarray
    waypoint: np.ndarray
    siteID: str

def read_data_file(data_filename):
    wifi = []
    waypoint = []
    siteID = ''

    with open(data_filename, 'r', encoding='utf-8') as file:
        lines = file.readlines()

    for line_data in lines:
        
        if 'SiteID:' in line_data:
            siteID = line_data.split('\t')[1].replace('SiteID:', '')

        line_data = line_data.strip()
        if not line_data or line_data[0] == '#':
            continue

        line_data = line_data.split('\t')
        
        if line_data[1] == 'TYPE_WAYPOINT':
            waypoint.append([int(line_data[0]), float(line_data[2]), float(line_data[3])])
            continue
        
        if line_data[1] == 'TYPE_WIFI':
            sys_ts = line_data[0]
            ssid = line_data[2]
            bssid = line_data[3]
            rssi = line_data[4]
            lastseen_ts = line_data[6]
            wifi_data = [sys_ts, ssid, bssid, rssi, lastseen_ts]
            wifi.append(wifi_data)
            continue
    
    wifi = np.array(wifi)
    waypoint = np.array(waypoint)
    #print(wifi)
    
    return ReadData(wifi, waypoint, siteID)

def toFloat(siteID):
    i = 0
    count = 0
    while i < len(siteID):
        count += ord(siteID[i])
        i += 1
    return float(count) 
    
sample_file = read_data_file("/Users/neel2sharma/indoor-location-navigation/train/5a0546857ecc773753327266/F2/5dccf516c04f060006e6e3c9.txt")

print('wifi shape:', sample_file.wifi.shape)
print('waypoint shape:', sample_file.waypoint.shape)
```

```python
z = 0
all_timestamps = []
all_dfs = {}
for building in used_buildings:

    folders = sorted(glob.glob(os.path.join('/Users/neel2sharma/indoor-location-navigation/','train/'+building+'/*')))
    print(building)

    small_train_x = []
    train_yx = []
    train_yy = []
    train_yz = []
    train_xfloor = []
    siteDict = {}
    curr_site = ""

    for folder in folders:

        floor = floor_map[folder.split('/')[-1]]
        files = glob.glob(os.path.join(folder, "*.txt"))
        
        for file in files:
            
            train_file = read_data_file(file)

            curr_site = train_file.siteID

            #since siteID is unique, converted int should be unique too
            #convert to int, append to map (int: str) and to train_xfloor
            #use map to convert predictions later
            key = toFloat(train_file.siteID)
            siteDict[key] = train_file.siteID

            train_xfloor.append(key)

            waypoint_values_x = train_file.waypoint[:,1]
            waypoint_values_y = train_file.waypoint[:,2]
            waypoint_timestamps = train_file.waypoint[:,0]

            #Append x and y data for this file into train_yx and train_yy
            i = 0
            while i < len(waypoint_values_x):
                train_yx.append(waypoint_values_x[i].astype(int))
                i += 1
                
            i = 0
            while i < len(waypoint_values_y):
                train_yy.append(waypoint_values_y[i].astype(int))
                train_yz.append(floor)
                i += 1
                
            #Select closest timestamps
            wifi_timestamps = [0] * len(waypoint_timestamps)
            i = 0
            while i < len(waypoint_timestamps):
                for x in train_file.wifi:
                    y = x[0].astype(float)
                    if abs(waypoint_timestamps[i] - y) < abs(waypoint_timestamps[i] - wifi_timestamps[i]):
                        wifi_timestamps[i] = y
                i += 1

            #now that closest timestamps have been selected, add them to a larger running list and create a df of rssi values
            #then add df to larger df (which is train_x)
            for x in wifi_timestamps:
                all_timestamps.append(x)

            bssids = []
            lists = []
            for x in train_file.wifi:
                if x[2] not in bssids:
                    bssids.append(x[2])
                    toAppend = [0] * len(wifi_timestamps)
                    lists.append(toAppend)

            i = 0
            while i < len(wifi_timestamps):
                for x in train_file.wifi:
                    if wifi_timestamps[i].astype(float) == x[0].astype(float):
                        if x[2] in bssids:
                            if x[3] == 'nan':
                                lists[bssids.index(x[2])][i] = 0
                            else:
                                lists[bssids.index(x[2])][i] = x[3].astype(int)
                i += 1
            
            i = 0
            while i < len(wifi_timestamps):
                wifi_timestamps[i] = wifi_timestamps[i] * (10 ** -12)
                i += 1
                
            file_df = pd.DataFrame({'timestamp': wifi_timestamps})
            i = 0
            while i < len(bssids):
                file_df[bssids[i].astype(str)] = lists[i]
                i += 1
            #display(file_df)
            small_train_x.append(file_df)



    train_x = pd.concat(small_train_x, ignore_index=True)
    train_x = train_x.fillna(0)
    #display(train_x)

    #here, use train_x, train_yx, and train_yy to fit RFX[z] and RFY[z]
    RFXS[curr_site].fit(train_x, train_yx)
    RFYS[curr_site].fit(train_x, train_yy)
    all_dfs[curr_site] = train_x
    RFZS[curr_site].fit(train_x, train_yz)
    #print('here')
    z += 1
```
